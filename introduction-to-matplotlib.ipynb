```python
# A matplotlib workflow

# 0. Import and get matplotlib ready
# %matplotlib inline # Not necessary in newer versions of Jupyter (e.g. 2022 onwards)
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot (Figure and Axes)
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Sample Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show
fig.savefig("images/simple-plot.png")
```

## Making figures with NumPy arrays

```python
import numpy as np
```

```python
# Create some data
x  = np.linspace(0,10,100)
x
```

```python
# Plot the data 
fig , ax = plt.subplots()
ax.plot(x,x**2);
```

We Want :

* Line Plot
* Scatter Plot
* Bar Plot
* Histogram
* Subplots

```python
# Use same data to make a scatter

fig , ax = plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
# Another scatter plot
fig , ax = plt.subplots()

ax.scatter(x,np.sin(x));
```

```python
# make some plot from dictionary
nut_butter_prices = {
    'Almond Bitter' : 10,
    'Ceshew' : 8,
    'peanut Butter' : 12
}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title='Vivek Nut Butter Store', ylabel='Prices (RS)');
```

```python
fig, ax = plt.subplots()
ax.barh(nut_butter_prices.keys(), nut_butter_prices.values());
```

```python
# Make some data from a normal distribution
x = np.random.randn(1000)

fig, ax = plt.subplots()
ax.hist(x)
```

Creating Figures with multiple Axes with Subplots

```python
fig , ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2, 
                                             figsize=(10, 5))
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```
